Outra maneira de criar uma tabela temporaria de uma tabela ja existente no FileStore para ser usada em SQL

DROP TABLE IF EXISTS nometabela - apaga caso exista

```
%sql
DROP TABLE IF EXISTS tabela_covid;

CREATE TABLE tabela_covid USING CSV
OPTIONS (path "/FileStore/tables/arquivo_geral.csv",header="true",delimiter=';')
```

```
%sql
SHOW TABLES
```

```
%sql
SELECT *
FROM tabela_covid
```
